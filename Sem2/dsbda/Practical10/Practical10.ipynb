```python
import docx2txt
import seaborn as sns
```

```python
path = 'testdoc.docx'
text = docx2txt.process(path)
```

```python
print(text)
```

```python
path2 = 'testdoc2.docx'
text2 = docx2txt.process(path2)
print(text2)
```

```python
#sentence tokenization
from nltk.tokenize import sent_tokenize
sent1 = sent_tokenize(text)
sent2 = sent_tokenize(text2)
print(sent1,sent2)
```

```python
#word tokenization
from nltk.tokenize import word_tokenize
word1 = word_tokenize(text)
word2 = word_tokenize(text2)
print(word1,word2)
```

```python
#to check frequency distribution of words
from nltk.probability import FreqDist
freq1 = FreqDist(word1)
freq2 = FreqDist(word2)
print(freq1)
print(freq2)
```

```python
#most common words in document 1
freq1.most_common()
```

```python
freq2.most_common()
```

```python
import matplotlib.pyplot as plt
freq1.plot(20)
plt.show()
```

```python
freq2.plot(20)
plt.show()
```

```python
#now removing the stop words
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print (stop_words)
```

```python
#remove the stopwords
new_word1 = []
for i in word1:
    if i not in stop_words:
        new_word1.append(i)

print(new_word1)
```

```python
new_word2 = []
for i in word2:
    if i not in stop_words:
        new_word2.append(i)

print(new_word2)
```

```python
#stemming
#removing the suffixes
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
stem_words1 = []
for i in new_word1:
    stem_words1.append(ps.stem(i))
print("Old words")
print(new_word1)
print("\n\n\n\nNew Stemmed words")
print(stem_words1)
```

```python
stem_words2 = []
for i in new_word2:
    stem_words2.append(ps.stem(i))
print("Old words")
print(new_word2)
print("\n\n\n\nNew Stemmed words")
print(stem_words2)
```

```python
#Lemmatization
#converting word in its base form like better converted to good
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
wnl = WordNetLemmatizer()
```

```python
lem_words1 = []
for i in new_word1:
    lem_words1.append(wnl.lemmatize(i))
print("Old words")
print(new_word1)
print("\n\n\n\nNew Lemmatized words")
print(lem_words1)
```

```python
lem_words2 = []
for i in new_word2:
    lem_words2.append(wnl.lemmatize(i))
print("Old Words")
print(new_word1)
print("\n\n\n\nNew Lemmatized words")
print(lem_words2)
```

```python
words = [text,text2]
wordset = set()

for i in words:
    temp = i.split()
    for word in temp:
        wordset.add(word)

wordset
```

```python
import numpy as np
dict = {}

def IDF():
    for it in wordset:
        temp = 0
        
        for i in range(2):
            if it in words[i].split():
                temp += 1
        dict[it] = np.log10(2/temp)
        print(it,dict[it])
                
IDF()
```

```python
document = [text,text2]
dict = {}
for it in document:
    for word in it.split():
        if word in dict:
            dict[word]+=1
        else:
            dict[word] = 1

    for k,v in dict.items():
        dict[k] = v/len(it)

print('Term Frequency')
for k,v in dict.items():
    print(k,v)
    
    
```
